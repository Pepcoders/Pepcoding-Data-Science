**Read cvs**

```python
df = pd.read_csv('https://raw.githubusercontent.com/neylsoncrepalde/projeto_eda_covid/master/covid_19_data.csv')
df
```

```python
df.info()
```

```python
df.head(8)
```

**Drop col and row**

```python
df = df.rename(columns = {'ObservationDate': 'Date', 'Province/State': 'State', 'Country/Region': 'Country' })
```

```python
df
```

```python
df['Date'] = pd.to_datetime(df['Date'])
```

```python
df.info()
```

```python
df.drop(['SNo', 'Last Update'], axis=1, inplace = True)
df.head()
```

```python
df.groupby('Country')[['Confirmed', 'Deaths', 'Recovered']].sum().reset_index()
```

```python
df.groupby(['Country', 'Date'])[['Confirmed', 'Deaths', 'Recovered']].sum().reset_index()
```
