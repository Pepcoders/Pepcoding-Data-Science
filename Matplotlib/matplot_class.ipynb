```python
import matplotlib.pyplot as plt
```

```python
x = [3, 7]
y = [10, 20]
plt.plot(x, y)

plt.show()
```

```python
plt.plot(x, y, "D", )

plt.show()
```

```python
y = [10, 20, 49, 31, 17, 89]
plt.plot(y)
plt.show()
```

```python
#marker | line | color
plt.plot(y, "o-.r",)
```

```python
plt.plot(y, marker = "o", ms = 30, mfc = "red", mec = "b")
```

```python
plt.plot(y, linestyle= "-.", linewidth = 5, color = "red")
```

```python
x = [2, 4, 5, 6, 7, 8, 9]
y = [10, 20, 30, 50, 20, 40, 12]


plt.scatter(x, y)
```

```python



x = [2, 4, 5, 6, 7, 8, 9, 10]
y = [10, 20, 30, 50, 20, 40, 12, 13]
size = [500, 600, 300, 676, 898, 8977, 7836, 53465]
plt.scatter(x, y, s = size, alpha = 0.3, color = "red")
```

```python
plt.bar(x, y, color = "red")
plt.title("sale")
plt.ylabel("rev")
plt.xlabel("time")
```

```python
plt.barh(x, y, color = "b", height = 0.5)
```

```python
import numpy as np

x = np.random.normal(100, 10, 250)

plt.hist(x)
plt.show()
print(x)
```

```python
x = ["IND", "USA", "NZ", "CHI"]
y = [10, 30, 40, 20]


plt.pie(y, labels= x, startangle = 90, explode=[0.2, 0, 0.4,0], shadow=True)

plt.show()
```

```python
x = [10, 20, 30]
y = [3, 5, 7]

plt.subplot(1, 2, 1)
plt.plot(x, y)
plt.title("sale")
plt.show()

plt.subplot(1, 2, 2)
plt.pie(x)
plt.title("rev")
plt.show()
```
